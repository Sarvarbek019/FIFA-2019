* **Information about the player:**
   * Name - player's  full name 
   * Age - player's age
   * Club - name of the team the player is playing in
   * Nationality - name of the national team

   * Potential - rating that players can achieve

   * Jersey number - number on the shirt

   * Height - player's height
   * Weight - player's weight

   * Joined - the date when he joined the club and signed the contract
*  **Skills of the player:** 
   * Finishing - completion of the attack or accuracy of the strike.The ability to be in the right place and score a goal
   * Ball Control - this is possession of the ball and vision of the field

   * Volleys - it is a skill that makes a precise shot without getting or handling the ball
   * Crossing - canopies or shot from the flank of the field.
   * Sprint speed - the maximum initial speed,starts an attack through the edge or counter attack.
   * Stamina - performing as much as possible,while maintaining the effectiveness of technical and tactical actions until the end of the match
   * GKReflexes - reaction of the goalkeeper and long-range batting
