```python
#create function value_and_wage_conversion
def value_and_wage_conversion(Value):
    out = Value.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in Value:
        out = float(out.replace('K', ''))*1000
    return float(out)


reader['Value'] = reader['Value'].apply(lambda x: value_and_wage_conversion(x))
reader['Wage'] = reader['Wage'].apply(lambda x: value_and_wage_conversion(x))
```

We are changing the icon of a million and a kilo at the numbers. We know a million has 6 zeros and a kilo has 3 zeros and we multiply them
